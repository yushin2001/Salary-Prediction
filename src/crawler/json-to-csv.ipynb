```python
def addApplyData (data, name):
    if len(data) > 2:
        d = {}
        for k in data:
            if len(k) <= 2:
                item = data[str(k)]
                # print(item)
                d [item[name]] = item['percent']
        return d
    else:
        return 'None'
```

```python
jobName, custName, appearDate, jobDescription, jobDescriptionLen = [], [], [], [], []
jobCategory, salaryType, salaryMin, salaryMax, addressArea, manageResp = [], [], [], [], [], []
businessTrip, workPeriod, vacationPolicy, startWorkingDay, needEmp = [], [], [], [], []
workExp, edu, major, language, specialty = [], [], [], [], []
skill, other, weltag = [], [], []
applyCount, applyEdu, applyYear, applySex, applyExp, applyLang = [], [], [], [], [], []
applyMajor, applySkill, applyCert = [], [], []
industry, capital, empNo, environmentPic = [], [], [], []

for i in range(len(raw)):
    if len(raw[i]['jobData'].keys()) == 17:
        job = raw[i]
        jobName.append(job['jobData']['header']['jobName'])
        custName.append(job['jobData']['header']['custName'])
        appearDate.append(job['jobData']['header']['appearDate'])

        content = job['jobData']['jobDetail']['jobDescription']
        content = content.replace('\n', '').replace(' ', '').replace('\r', '')
        content = re.sub("\[.{0,12}\]","", content) # remove emoji
        content = re.sub("http://[a-zA-z./\d]*","", content) # remove link
        jobDescription.append(content)
        jobDescriptionLen.append(len(content))

        job_category = []
        for cat in job['jobData']['jobDetail']['jobCategory']:
            job_category.append(cat['description'])
        jobCategory.append(job_category)

        salaryType.append(job['jobData']['jobDetail']['salary'][:2])
        salaryMin.append(job['jobData']['jobDetail']['salaryMin'])
        salaryMax.append(job['jobData']['jobDetail']['salaryMax'])
        addressArea.append(job['jobData']['jobDetail']['addressArea'])
        manageResp.append(job['jobData']['jobDetail']['manageResp'])
        businessTrip.append(job['jobData']['jobDetail']['businessTrip'])
        workPeriod.append(job['jobData']['jobDetail']['workPeriod'])
        vacationPolicy.append(job['jobData']['jobDetail']['vacationPolicy'])
        startWorkingDay.append(job['jobData']['jobDetail']['startWorkingDay'])
        needEmp.append(job['jobData']['jobDetail']['needEmp'])
        workExp.append(job['jobData']['condition']['workExp'])
        edu.append(job['jobData']['condition']['edu'])

        # condition: major
        if len(job['jobData']['condition']['major']) > 0:
            major.append(job['jobData']['condition']['major'])
        else:
            major.append('None')

        # condition: langage
        if len(job['jobData']['condition']['language']) > 0:
            landict = {}
            for lan in job['jobData']['condition']['language']:
                landict [lan['language']] = lan['ability']
            language.append(landict)
        else:
            language.append('None')

        # condition: specialty
        if len(job['jobData']['condition']['specialty']) > 0:
            spec = []
            for sp in job['jobData']['condition']['specialty']:
                spec.append(sp['description'])
            specialty.append(spec)
        else:
            specialty.append('None')
        
        # condition: skill
        if len(job['jobData']['condition']['skill']) > 0:
            skill_list = []
            for sk in job['jobData']['condition']['skill']:
                skill_list.append(sk['description'])
            skill.append(skill_list)
        else:
            skill.append('None')


        other.append(job['jobData']['condition']['other'])

        # welfare tag
        if len(job['jobData']['welfare']['tag']) > 0:
            weltag.append(job['jobData']['welfare']['tag'])
        else:
            weltag.append('None')


        applyCount.append(job['applyData']['sex']['total'])

        # apply data
        applyEdu.append(addApplyData(job['applyData']['edu'], 'eduName'))
        applyYear.append(addApplyData(job['applyData']['yearRange'], 'yearRangeName'))
        applySex.append(addApplyData(job['applyData']['sex'], 'sexName'))
        applyExp.append(addApplyData(job['applyData']['exp'], 'expName'))
        applyLang.append(addApplyData(job['applyData']['language'], 'langName'))
        applyMajor.append(addApplyData(job['applyData']['major'], 'majorName'))
        applySkill.append(addApplyData(job['applyData']['skill'], 'skillName'))
        applyCert.append(addApplyData(job['applyData']['cert'], 'certName'))

        # cust data
        industry.append(job['jobData']['industry'])
        capital.append(job['custData']['capital'])
        empNo.append(job['custData']['empNo'])

        if len(job['jobData']['environmentPic']['environmentPic']) > 0:
            environmentPic.append(1)
        else:
            environmentPic.append(0)
```

```python
data_dict = {'jobName': jobName, 'custName': custName, 'appearDate': appearDate, 'jobDescription': jobDescription,
                    'jobDescriptionLen': jobDescriptionLen, 'jobCategory':jobCategory, 'salaryType': salaryType, 'salaryMin': salaryMin, 'salaryMax':salaryMax,
                    'addressArea': addressArea, 'manageResp':manageResp, 'businessTrip': businessTrip, 'workPeriod': workPeriod,
                    'vacationPolicy': vacationPolicy, 'startWorkingDay': startWorkingDay, 'needEmp': needEmp, 'workExp': workExp,
                    'edu': edu, 'major': major, 'language': language, 'specialty': specialty, 'skill': skill,
                    'other': other, 'weltag': weltag, 'applyCount': applyCount, 'applyEdu': applyEdu, 'applyYear': applyYear,
                    'applySex': applySex, 'applyExp': applyExp, 'applyLang': applyLang, 'applyMajor': applyMajor,
                    'applySkill': applySkill, 'applyCert': applyCert, 'industry': industry, 'capital': capital,
                    'empNo': empNo, 'environmentPic': environmentPic}

# for k, v in data_dict.items():
#     print(k, len(v))
```
