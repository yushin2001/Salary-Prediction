```python
import pandas as pd
import json
```

```python
df = pd.read_csv('./jobdata2.csv')
df
```

## 清除年薪的資料

```python
df2 = df[(df['salaryMax'] != 9999999) & (df['salaryMin'] > 10000)]
df2
```

```python
df_yr = df2[df2['salaryType'] == '年薪']
c = df_yr['salaryMin'].min()
d = df_yr['salaryMin'].max()
c2 = df_yr['salaryMax'].min()
d2 = df_yr['salaryMax'].max()

df_dy = df2[df2['salaryType'] == '日薪']
e = df_dy['salaryMin'].min()
f = df_dy['salaryMin'].max()
e2 = df_dy['salaryMax'].min()
f2 = df_dy['salaryMax'].max()

df_hr = df2[df2['salaryType'] == '時薪']
g = df_hr['salaryMin'].min()
h = df_hr['salaryMin'].max()
g2 = df_hr['salaryMax'].min()
h2 = df_hr['salaryMax'].max()

print(f'年薪 salaryMin range: {c},{d} | salaryMax range: {c2},{d2}')
print(f'日薪 salaryMin range: {e},{f} | salaryMax range: {e2},{f2}')
print(f'時薪 salaryMin range: {g},{h} | salaryMax range: {g2},{h2}')
```

```python
df3 = df2[(df2['salaryMax'] < 340000) & (df2['salaryMin'] < 303000)]
df4 = df3[df3['salaryType'] != '月薪']
df4
```

## EDA

#### salryMin & salaryMax

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

label = 'salaryMin', 'salaryMax'
dfmin = df3['salaryMin']
dfmax = df3['salaryMax']
plt.figure(figsize=(5,5))
plt.title('月薪盒狀圖（去除異常值）',fontsize=15)
plt.boxplot([dfmin, dfmax], patch_artist=True, boxprops = {'color':'orangered','facecolor':'lightyellow'}, labels=label, showfliers=False)
plt.show()
```

#### Frequency of area

```python
area_cat = list(df3['addressArea'].unique())
area = list(df3['addressArea'])
area_count = []

for cat in area_cat:
    count = 0
    for ar in area:
        if cat == ar: 
            count += 1
    area_count.append(count)
```

```python
plt.figure(figsize=(20,8))
plt.title('職缺的地區分佈',fontsize=20)
plt.bar(x=area_cat, height=area_count)
plt.xlabel('縣市/國家', fontsize=15)
plt.ylabel('數量', fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

#### salaryMin & area

```python
df_area = df3.loc[:, ['addressArea', 'salaryMin']]
df_area
```

```python
sMin = list(df_area['salaryMin'])
area_list = list(df_area['addressArea'])
area_dict = {}

for cat in area_cat:
    area_dict[cat] = []
    for i in range(len(sMin)):
        if cat == area_list[i]:
            area_dict[cat].append(sMin[i])
                
```

```python
boxlist = []

for k, v in area_dict.items():
    boxlist.append(v)

boxlist = boxlist[:8] # 只留下六都跟新竹

plt.figure(figsize=(10,5))
plt.title('六都+新竹的起薪盒狀圖（去除異常值）',fontsize=20)
plt.xlabel('縣市', fontsize=15)
plt.ylabel('起薪', fontsize=15)
plt.boxplot(boxlist, patch_artist=True, showfliers=False, labels=area_cat[:8], boxprops = {'color':'orangered','facecolor':'peachpuff'})
plt.xticks(rotation=45)
plt.show()
```

#### 起薪Top10與Last 10的產業（by 中位數）

```python
df_ind = df3.loc[:, ['industry', 'salaryMin']]
df_ind
```

```python
from statistics import median

ind_cat = list(df_ind['industry'].unique())
ind_list = list(df_ind['industry'])
sMin = list(df_ind['salaryMin'])

ind_dict = {}
med_dict = {}
for cat in ind_cat:
    ind_dict[cat] = []
    med_dict[cat] = 0
    for i in range(len(sMin)):
        if cat == ind_list[i]:
            ind_dict[cat].append(sMin[i])
    med = median(ind_dict[cat])
    med_dict[cat] = med
```

```python
len(ind_cat)
```

```python
sort_med_dict = dict(sorted(med_dict.items(), key=lambda item: item[1], reverse=True)) # decreasing order
topind = []
lastind = []
toplist = []
lastlist = []

index = 1
for k, v in sort_med_dict.items():
    if index <= 10:
        topind.append(ind_dict[k])
        toplist.append(k)
    elif index > 260:
        lastind.append(ind_dict[k])
        lastlist.append(k)
    index += 1
```

```python
plt.figure(figsize=(15,5))
plt.title('起薪Top10產業',fontsize=20)
plt.xlabel('產業', fontsize=15)
plt.ylabel('起薪', fontsize=15)
plt.boxplot(topind, patch_artist=True, showfliers=False, labels=toplist, boxprops = {'color':'orangered','facecolor':'pink'})
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('起薪Last10產業',fontsize=20)
plt.xlabel('產業', fontsize=15)
plt.ylabel('起薪', fontsize=15)
plt.boxplot(lastind, patch_artist=True, showfliers=False, labels=lastlist, boxprops = {'color':'orangered','facecolor':'lightsteelblue'})
plt.xticks(rotation=45)
plt.show()
```
